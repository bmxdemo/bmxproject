```python
import numpy as np
from numpy.fft import rfft,irfft
%pylab inline
```

```python
Na=2048 # let's have Nachunks
Nx=2048 # of Nx samples for a total sample
N=Na*Nx #roughly milisecond
dt=1.0/2.5e9 #2.5 GSample/s
t=arange(N)*dt
tmax=t[-1]
print "total sample = ",tmax/1e-3,"ms"
```

```python
# generate sample
# we will generate tone at 1MHz, 102MHz, 203MHz, etc up to 1213MHz
sam=zeros(N)
for f in arange(1,1214,101):
    phi=np.random.uniform(0,2*np.pi)
    print "Adding ",f,'MHz'
    sam+=sin(t*2*pi*f*1e6+phi)
```

```python
## first the total fft
print "Doing fft"
ft=rfft(sam)
freq=arange(len(ft))/tmax/1e6 ## In Mhz
## average
fta,freqa=[],[]
print "Averaging"
# averaging
for i in range(Na/2):
    tm=ft[i*Nx:(i+1)*Nx]
    fta.append(abs(tm*tm).mean())
    freqa.append(freq[i*Nx:(i+1)*Nx].mean())
    
```

```python
plot(freqa,fta)
ma=array(fta).max()
for f,a in zip(freqa,fta):
    if a/ma>0.5:
        print f,a/ma
```

```python
## now get the kernel convolution kernel
fkern=np.ones(len(freq),dtype=complex)
fkern[np.where(freq<1000)]=0.0
rkern=irfft(fkern)
tx=t[0:40]
figure()
plot(tx,rkern[0:40])
plot(tx,-1/np.sqrt(2)*sin(tx*1000e6*2*np.pi)/(tx*800e6*2*np.pi))
## now truncate at 10 samples and invert and plot
figure()
KS=50
rkern[KS+1:-KS-1]=0
fkernt=rfft(rkern)
plot(freq,fkernt**2)
```

```python
print arange(KS,0,-1)
sam2=zeros(N)
for i in arange(KS,0,-1):
    sam2[i:]+=+rkern[i]*sam[:-i]
    sam2[:-i]+=+rkern[-i]*sam[i:]
sam2+=rkern[0]*sam
print "FFT"
ft2=rfft(sam2)
fta2=[]
print "average"
for i in range(Na/2):
    tm=ft2[i*Nx:(i+1)*Nx]
    fta2.append(abs(tm*tm).mean())
fta2=array(fta2)
plot(0+array(freqa),fta2/fta2.max())

semilogy()
```

```python
sam3=sam2*sin(1000e6*2*np.pi*t)
ft3=rfft(sam3)
fta3=[]
print "average"
for i in range(Na/2):
    tm=ft3[i*Nx:(i+1)*Nx]
    fta3.append(abs(tm*tm).mean())
fta3=array(fta3)
plot(1000+array(freqa),fta3/fta3.max())
xlim(1000,1250)
semilogy()
```

```python
## now see if we can undersample by a factor of 3 and still get the same FFT
sam4=sam3[:-1:3]+sam3[1::3]+sam3[2::3]
ft4=rfft(sam4)
fta4=[]
freqa4=[]
print "average"
for i in range(Na/2/3):
    tm=ft4[i*Nx:(i+1)*Nx]
    fta4.append(abs(tm*tm).mean())
    #freqa4.append(freq[i*Nx/3:(i+1)*Nx/3].mean())
    
fta4=array(fta4)
plot(1000+array(freqa[:len(fta4)]),fta4/fta4.max())
xlim(1000,1250)
semilogy()
```
