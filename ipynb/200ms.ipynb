```python
import numpy as np
from numpy.fft import rfft,irfft
%pylab inline
```

```python
realdata=np.fromfile("../../data/200ms_500MHz.bin",'u2')
```

```python
N=len(realdata)
mean=realdata.mean()
rms=np.sqrt(realdata.var())
Ns=4096
Na=131835
print mean, rms
print N,Na*Ns
dt=1.0/2.7e9 #2.5 GSample/s
nuX=500.69158364*1e6
```

```python
# first let's check weird ordering, by taking firt 100 samples
figure (figsize=(16,12))
t=arange(100)*dt
y=realdata[:100]
plot(t,y,'ro--')
plot(t,mean+np.sqrt(2)*rms*(sin(2+nuX*t*np.pi*2)))
```

```python
fd=fft.rfft(realdata)
```

```python
def plotlocal(nu1,nu2):
    dnu=1.0/(dt*N)
    i1=int(nu1/dnu)
    i2=int(nu2/dnu)
    far=arange(i1,i2)*dnu
    fdz=fd[i1:i2]
    fdz=abs(fdz*fdz)
    figure(figsize=(16,12))
    plot(far/1e6,fdz)
    xlabel('f[MHz]')
    ylabel('power')
    semilogy()
    return far,fdz
```

```python
far,fdz=plotlocal(500.6e6,500.8e6)
```

```python
nuM=far[fdz.argmax()]
dnu=1/(N*dt)
print nuM/1e6, "+/-", dnu/1e6
print "ratio",nuM/nuX-1
```

```python
## Do we still see the 6.75MHz
far,fdz=plotlocal(670e6,680e6)
```

```python
## What about 1/f?
far,fdz=plotlocal(10,10e6)
```

```python
Navg=250000
dnua=1.0/(dt*N)*Navg
i=1
res=[]
while i<N/2:
    tmp=fd[i:i+Navg]
    res.append(abs(tmp*tmp).mean())
    i+=Navg
pow=np.array(res)
freq=arange(len(res))*dnua
```

```python
figure(figsize=(16,12))
plot(freq/1e6,pow,'ro-')
xlabel("f[MHz]")
ylabel("power")
semilogy()
#xlim(990,1050)
print len(freq)
```
